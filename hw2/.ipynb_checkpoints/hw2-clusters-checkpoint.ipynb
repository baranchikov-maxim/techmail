Алгоритм кластеризации должен удовлетворять следующему интерфейсу. Конструктор принимает набор параметров, необходимых для работы алгоритма кластеризации. Метод `fit` подсчитывает параметры модели и возвращает `self`. Метод `predict` возвращает вектор с индексами кластеров для поданных в него объектов `x`

```python
%%cython

import numpy as np
cimport numpy as np
import sklearn.metrics as smt

from sklearn.base import BaseEstimator, ClusterMixin
        
    
cpdef double single(list cl1, list cl2, np.ndarray[np.float64_t, ndim=2] metrics):
    cdef:
        double min=float('+inf'), dist
        int i, j, len_cl1 = len(cl1), len_cl2 = len(cl2)
    for i in xrange(len_cl1):
        for j in xrange(len_cl2):
            dist = metrics[cl1[i], cl2[j]]
            if dist < min:
                min = dist
    return min

cpdef double complete(list cl1, list cl2, np.ndarray[np.float64_t, ndim=2] metrics):
    cdef:
        double max=float('-inf'), dist
        int i, j, len_cl1 = len(cl1), len_cl2 = len(cl2)
    for i in xrange(len_cl1):
        for j in xrange(len_cl2):
            dist = metrics[cl1[i], cl2[j]]
            if dist > max:
                max = dist
    return max

cpdef double average(list cl1, list cl2, np.ndarray[np.float64_t, ndim=2] metrics):
    cdef:
        double average=0, dist
        int i, j, len_cl1 = len(cl1), len_cl2 = len(cl2)
    for i in xrange(len_cl1):
        for j in xrange(len_cl2):
            average += metrics[cl1[i], cl2[j]]
    return average / len_cl1 / len_cl2


class Clustering(BaseEstimator, ClusterMixin):
    
    def __init__(self, k, linkage='single'):
        self.k = k
        self.linkage_metric = Clustering.get_linkage_metric(linkage)
        self.labels = None
        
        
    def fit(self, x, y=None):
        x = np.array(x) if type(x) is not type(np.ndarray) else x
        x_metrics = np.array(smt.pairwise.pairwise_distances(x))
        clusters = [[i] for i in xrange(len(x))]
                
        while len(clusters) > self.k:
            # to_be_merged is a pair of cluster indices
            to_be_merged = self.find_nearest_clusters(clusters, x_metrics)
            
            clusters[to_be_merged[0]] += clusters[to_be_merged[1]]
            del clusters[to_be_merged[1]]
            
            
        self.labels = np.empty(x.shape[0], dtype=np.dtype('int'))
        for num, cluster in enumerate(clusters):
            self.labels[cluster] = num
            
        return self
    
    def predict(self, x):
        return None
    
    def fit_predict(self, x, y=None):
        self.fit(x, y)
        return self.predict(x)
    
    def find_nearest_clusters(self, clusters, metrics):
        cdef double min_dist = float('+inf'), dist
        cdef int min_i, min_j, i, j, _len = len(clusters)
        for i in range(_len):
            for j in range(i+1, _len):
                dist = self.linkage_metric(clusters[i], clusters[j], metrics)
                if dist < min_dist:
                    min_i = i
                    min_j = j
                    min_dist = dist
                    
        return (min_i, min_j)
    
    @staticmethod
    def get_linkage_metric(linkage_name):
        kw = {'single': single, 'complete': complete, 'average': average}
        if linkage_name not in kw:
            raise ValueError('Unknow linkage: %s' % linkage_name)
        return kw[linkage_name]
        
```

Выбрать и реализовать критерий качества, на основании которого будут подбираться параметры модели. Варианты критериев:
- Инерция (сумма квадратов расстояний от каждой из точек до ближайшего к этой точке центроида)
- Средний диаметр (максимальное расстояние между точками в кластере) или радиус (расстояние от центроида до самой дальней от него точки в кластере)
- Silhouette

**Критерий качества необходимо выбрать таким образом, чтобы он подходил для реализованного алгоритма**

## В общем случае в зависимости от типа связывания могут подходить разные критерии.
Рассмотрим single linkage - вариант когда кластер "проглатывет" ближайшие точки. В этом случае ожидается, что точки в кластере будут в среднем ближе к друг другу, нежели к точкам другого кластера, т.е. критерий - silhouette

```python
def quality(x, y):
    def mean_in_cluster(x, y, idx):
        indices = np.where(y == y[idx])[0]
        return np.mean([np.linalg.norm(x[i] - x[j]) for j in indices if not i == j])
    
    def mean_outside(x, y, idx):
        return np.min(
            [np.mean([np.linalg.norm(x[i] - x[j]) for j in np.where(y == cur_label)[0]]
            ) for cur_label in np.unique(y) if not cur_label == y[idx]])
    
    a = np.array([mean_in_cluster(x, y, i)
                  for i in xrange(len(y))])
    b = np.array([mean_outside(x, y, i)
                  for i in xrange(len(y))])
    
    return np.mean((b-a)/np.maximum(a,b))
```

Применим критерий для выбора параметров алгоритма. Предлагается изменить следующий код таким образом, чтобы кластеризация вызывалась с верным набором параметров. На графике можно будет увидеть зависимость критерия от параметров алгоритма. Необходимо выбрать оптимальные значения этих параметров.

```python
def plot_quality(x, maxk, linkage='single'):
    ks = range(2, maxk)
    criteria = np.zeros(len(ks))

    for i, k in enumerate(ks):
        y = Clustering(k, linkage=linkage).fit(x).labels
        criteria[i] = quality(x, y)
    
    pl.figure(figsize=(8, 6))
    pl.plot(ks, criteria)
    pl.title("$J(k)$")
    pl.ylabel("Criteria $J$")
    pl.xlabel("Number of clusters $k$")
    pl.grid()
    pl.show()
```

```python
plot_quality(x, 30, 'average')
```
